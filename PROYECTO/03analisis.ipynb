# An치lisis exploratorio completo del dataset 1DATABASE.csv

```python
import pandas as pd #Inspecci칩n general de filas y columnas

# Cargar archivo CSV
df = pd.read_csv("1DATABASE1.csv", sep=';')
df.head()
```

```python
#Informaci칩n general del dataset
print("Dimensiones (filas, columnas):", df.shape)
df.info()
```

```python
#Reemplazo de comas por puntos en notas
df['Nota 1'] = df['Nota 1'].astype(str).str.replace(',', '.').astype(float)
df['Nota 2'] = df['Nota 2'].astype(str).str.replace(',', '.').astype(float)
df.head()
```

```python
#Estad칤stica descriptiva general
df.describe(include='all')
```

```python
#Verificaci칩n de valores nulos y duplicados
print("Valores nulos por columna:\n", df.isnull().sum())
print("\nFilas duplicadas:", df.duplicated().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Variables num칠ricas
numeric_cols = ['Nota 1', 'Nota 2', 'Asistencia Clases (%)', 'Asistencia Laboratorio (%)']

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=10)
    plt.title(f'Distribuci칩n de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.grid(True)
    plt.show()
```

```python
# Boxplots para detecci칩n de valores at칤picos
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot de {col}')
    plt.grid(True)
    plt.show()
```

```python
# Matriz de correlaci칩n
plt.figure(figsize=(8,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Matriz de Correlaci칩n entre Variables Num칠ricas")
plt.show()
```
